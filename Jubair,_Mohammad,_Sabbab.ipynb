```python
import urllib.request
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX_South_Texas_launch_site')
html = response.read()
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
```

```python
print(text)
```

```python
tokens = [t for t in text.split()]
```

```python
print(tokens)
```

```python
import nltk
freq_before = nltk.FreqDist(tokens)

#for key,val in freq.items():

    #print (str(key) + ':' + str(val))
```

```python
freq_before.plot(10,cumulative=False)
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
clean_tokens = tokens[:]

sr = stopwords.words('english')

for token in tokens:
  if token in stopwords.words('english'):
    clean_tokens.remove(token)
  
print(clean_tokens)
freq_after = nltk.FreqDist(clean_tokens)
```

```python
freq_after.plot(10,cumulative=False)
```

```python
# Ans for Question 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

all_fdist = nltk.FreqDist(freq_after).most_common(10)

all_fdist = pd.Series(dict(all_fdist))

fig, ax = plt.subplots(figsize=(10,10))

all_plot = sns.barplot(x=all_fdist.index, y=all_fdist.values, ax=ax)
plt.xticks(rotation=30);
```
